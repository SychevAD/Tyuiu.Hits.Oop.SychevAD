```
using System;
using System.Collections.Generic;

// Базовый интерфейс для дополнительного функционала
public interface IValidatable
{
    bool Validate();
}

public interface IExportable
{
    string ExportToJson();
    string ExportToXml();
}

// Базовый класс для всех услуг (для множественного наследования)
public abstract class Service
{
    private string serviceId;
    private DateTime createdDate;
    private string status;

    public string ServiceId
    {
        get { return serviceId; }
        set { serviceId = value; }
    }

    public DateTime CreatedDate
    {
        get { return createdDate; }
        set { createdDate = value; }
    }

    public string Status
    {
        get { return status; }
        set { status = value; }
    }

    public abstract void DisplayServiceInfo();
    
    public virtual void ActivateService()
    {
        Console.WriteLine($"Услуга {ServiceId} активирована");
    }
    
    protected Service(string serviceId)
    {
        ServiceId = serviceId;
        CreatedDate = DateTime.Now;
        Status = "Active";
    }
}

public abstract class Ticket : IValidatable, IExportable
{
    private decimal price;
    private string eventName;
    private DateTime eventDate;
    private string venue;

    // Новые атрибуты
    private string ticketId;
    private string barcode;
    private bool isVerified;
    private int loyaltyPoints;

    public string TicketId
    {
        get { return ticketId; }
        set { ticketId = value; }
    }

    public string Barcode
    {
        get { return barcode; }
        set { barcode = value; }
    }

    public bool IsVerified
    {
        get { return isVerified; }
        set { isVerified = value; }
    }

    public int LoyaltyPoints
    {
        get { return loyaltyPoints; }
        set { loyaltyPoints = value; }
    }

    public string EventName
    {
        get { return eventName; }
        set { eventName = value; }
    }

    public DateTime EventDate
    {
        get { return eventDate; }
        set { eventDate = value; }
    }

    public string Venue
    {
        get { return venue; }
        set { venue = value; }
    }

    public decimal Price
    {
        get { return price; }
        set { price = value; }
    }
    
    // Новые методы
    public virtual bool Validate()
    {
        return !string.IsNullOrEmpty(EventName) && Price > 0 && EventDate > DateTime.Now;
    }
    
    public virtual string ExportToJson()
    {
        return $"{{\"eventName\":\"{EventName}\",\"price\":{Price},\"date\":\"{EventDate:yyyy-MM-dd}\"}}";
    }
    
    public virtual string ExportToXml()
    {
        return $"<Ticket><EventName>{EventName}</EventName><Price>{Price}</Price><Date>{EventDate:yyyy-MM-dd}</Date></Ticket>";
    }
    
    public void VerifyTicket()
    {
        IsVerified = true;
        Console.WriteLine($"Билет {TicketId} верифицирован");
    }
    
    public int CalculateLoyaltyPoints()
    {
        LoyaltyPoints = (int)(Price * 0.1m);
        return LoyaltyPoints;
    }
    
    public abstract void DisplayTicketInfo();
    
    public virtual void Reserve()
    {
        Console.WriteLine($"Билет на {EventName} забронирован");
    }
    
    protected Ticket(string eventName, DateTime eventDate, string venue, decimal price)
    {
        EventName = eventName;
        EventDate = eventDate;
        Venue = venue;
        Price = price;
        TicketId = Guid.NewGuid().ToString();
        Barcode = $"BC_{DateTime.Now.Ticks}";
        IsVerified = false;
        LoyaltyPoints = 0;
    }
}

// Простое наследование
public class MovieTicket : Ticket
{
    private string movieTitle;
    private int hallNumber;
    private string seatNumber;

    // Новые атрибуты
    private string genre;
    private int durationMinutes;
    private string language;
    private bool is3D;

    public string Genre
    {
        get { return genre; }
        set { genre = value; }
    }

    public int DurationMinutes
    {
        get { return durationMinutes; }
        set { durationMinutes = value; }
    }

    public string Language
    {
        get { return language; }
        set { language = value; }
    }

    public bool Is3D
    {
        get { return is3D; }
        set { is3D = value; }
    }

    public string MovieTitle
    {
        get { return movieTitle; }
        set { movieTitle = value; }
    }

    public int HallNumber
    {
        get { return hallNumber; }
        set { hallNumber = value; }
    }

    public string SeatNumber
    {
        get { return seatNumber; }
        set { seatNumber = value; }
    }
    
    // Новые методы
    public void DisplayMovieDetails()
    {
        Console.WriteLine($"Жанр: {Genre}, Продолжительность: {DurationMinutes} мин.");
        Console.WriteLine($"Язык: {Language}, 3D: {(Is3D ? "Да" : "Нет")}");
    }
    
    public TimeSpan GetRemainingTime()
    {
        return EventDate - DateTime.Now;
    }
    
    public override bool Validate()
    {
        return base.Validate() && !string.IsNullOrEmpty(MovieTitle) && HallNumber > 0;
    }
    
    public override void DisplayTicketInfo()
    {
        Console.WriteLine($"КИНОБИЛЕТ: {MovieTitle}");
        Console.WriteLine($"Дата: {EventDate:dd.MM.yyyy HH:mm}");
        Console.WriteLine($"Зал: {HallNumber}, Место: {SeatNumber}");
        Console.WriteLine($"Цена: {Price} руб.");
        Console.WriteLine($"Место проведения: {Venue}");
        DisplayMovieDetails();
    }
    
    public override void Reserve()
    {
        base.Reserve();
        Console.WriteLine($"Место {SeatNumber} в зале {HallNumber} зарезервировано");
    }
    
    public void WatchTrailer()
    {
        Console.WriteLine($"Просмотр трейлера к фильму '{MovieTitle}'");
    }
    
    public MovieTicket(string movieTitle, DateTime eventDate, string venue, 
                      decimal price, int hallNumber, string seatNumber)
        : base(movieTitle, eventDate, venue, price)
    {
        MovieTitle = movieTitle;
        HallNumber = hallNumber;
        SeatNumber = seatNumber;
        Genre = "Не указан";
        DurationMinutes = 120;
        Language = "Русский";
        Is3D = false;
    }
}

// Сложное наследование с переопределением методов
public class ConcertTicket : Ticket
{
    private string artist;
    private string ticketType;
    private bool hasBackstageAccess;

    // Новые атрибуты
    private string musicGenre;
    private int expectedDuration;
    private string specialEffects;
    private bool meetAndGreet;

    public string MusicGenre
    {
        get { return musicGenre; }
        set { musicGenre = value; }
    }

    public int ExpectedDuration
    {
        get { return expectedDuration; }
        set { expectedDuration = value; }
    }

    public string SpecialEffects
    {
        get { return specialEffects; }
        set { specialEffects = value; }
    }

    public bool MeetAndGreet
    {
        get { return meetAndGreet; }
        set { meetAndGreet = value; }
    }

    public string Artist
    {
        get { return artist; }
        set { artist = value; }
    }

    public string TicketType
    {
        get { return ticketType; }
        set { ticketType = value; }
    }

    public bool HasBackstageAccess
    {
        get { return hasBackstageAccess; }
        set { hasBackstageAccess = value; }
    }
    
    // Новые методы
    public void DisplayConcertDetails()
    {
        Console.WriteLine($"Музыкальный жанр: {MusicGenre}");
        Console.WriteLine($"Продолжительность: {ExpectedDuration} минут");
        Console.WriteLine($"Спецэффекты: {SpecialEffects}");
        Console.WriteLine($"Встреча с артистом: {(MeetAndGreet ? "Да" : "Нет")}");
    }
    
    public override int CalculateLoyaltyPoints()
    {
        LoyaltyPoints = (int)(Price * 0.15m); // Больше баллов за концерты
        if (HasBackstageAccess) LoyaltyPoints += 50;
        return LoyaltyPoints;
    }
    
    public override void DisplayTicketInfo()
    {
        Console.WriteLine($"КОНЦЕРТНЫЙ БИЛЕТ: {Artist}");
        Console.WriteLine($"Дата: {EventDate:dd.MM.yyyy HH:mm}");
        Console.WriteLine($"Тип билета: {TicketType}");
        Console.WriteLine($"Доступ за кулисы: {(HasBackstageAccess ? "Да" : "Нет")}");
        Console.WriteLine($"Цена: {Price} руб.");
        Console.WriteLine($"Место проведения: {Venue}");
        DisplayConcertDetails();
    }
    
    public override void Reserve()
    {
        Console.WriteLine($"Билет на концерт {Artist} класса {TicketType} забронирован");
        if (HasBackstageAccess)
        {
            Console.WriteLine("Включен доступ за кулисы!");
        }
        if (MeetAndGreet)
        {
            Console.WriteLine("Запланирована встреча с артистом!");
        }
    }
    
    public void SoundCheck()
    {
        Console.WriteLine($"Проверка звука перед концертом {Artist}");
    }
    
    public ConcertTicket(string artist, DateTime eventDate, string venue,
                        decimal price, string ticketType, bool backstageAccess)
        : base($"{artist} Concert", eventDate, venue, price)
    {
        Artist = artist;
        TicketType = ticketType;
        HasBackstageAccess = backstageAccess;
        MusicGenre = "Разнообразный";
        ExpectedDuration = 120;
        SpecialEffects = "Стандартные";
        MeetAndGreet = false;
    }
}

// Множественное наследование через интерфейсы
public class TheaterTicket : Ticket, IExportable
{
    private string playName;
    private string director;
    private int actNumber;

    // Новые атрибуты
    private string playwright;
    private int intermissionCount;
    private string stageType;
    private bool hasLiveMusic;

    public string Playwright
    {
        get { return playwright; }
        set { playwright = value; }
    }

    public int IntermissionCount
    {
        get { return intermissionCount; }
        set { intermissionCount = value; }
    }

    public string StageType
    {
        get { return stageType; }
        set { stageType = value; }
    }

    public bool HasLiveMusic
    {
        get { return hasLiveMusic; }
        set { hasLiveMusic = value; }
    }

    public string PlayName
    {
        get { return playName; }
        set { playName = value; }
    }

    public string Director
    {
        get { return director; }
        set { director = value; }
    }

    public int ActNumber
    {
        get { return actNumber; }
        set { actNumber = value; }
    }
    
    // Новые методы
    public void DisplayTheaterDetails()
    {
        Console.WriteLine($"Драматург: {Playwright}");
        Console.WriteLine($"Количество антрактов: {IntermissionCount}");
        Console.WriteLine($"Тип сцены: {StageType}");
        Console.WriteLine($"Живая музыка: {(HasLiveMusic ? "Да" : "Нет")}");
    }
    
    public override string ExportToJson()
    {
        return $"{{\"playName\":\"{PlayName}\",\"director\":\"{Director}\",\"price\":{Price},\"date\":\"{EventDate:yyyy-MM-dd}\"}}";
    }
    
    public override string ExportToXml()
    {
        return $"<TheaterTicket><PlayName>{PlayName}</PlayName><Director>{Director}</Director><Price>{Price}</Price><Date>{EventDate:yyyy-MM-dd}</Date></TheaterTicket>";
    }
    
    public override void DisplayTicketInfo()
    {
        Console.WriteLine($"ТЕАТРАЛЬНЫЙ БИЛЕТ: {PlayName}");
        Console.WriteLine($"Режиссер: {Director}");
        Console.WriteLine($"Дата: {EventDate:dd.MM.yyyy HH:mm}");
        Console.WriteLine($"Акт: {ActNumber}");
        Console.WriteLine($"Цена: {Price} руб.");
        Console.WriteLine($"Театр: {Venue}");
        DisplayTheaterDetails();
    }
    
    public void ReadSynopsis()
    {
        Console.WriteLine($"Чтение синопсиса пьесы '{PlayName}'");
    }
    
    public TheaterTicket(string playName, string director, DateTime eventDate,
                        string venue, decimal price, int actNumber)
        : base(playName, eventDate, venue, price)
    {
        PlayName = playName;
        Director = director;
        ActNumber = actNumber;
        Playwright = "Неизвестен";
        IntermissionCount = 1;
        StageType = "Стандартная";
        HasLiveMusic = false;
    }
}

// Класс для демонстрации множественного наследования
public class VIPTicket : TheaterTicket, IExportable
{
    // Новые атрибуты для VIP
    private string vipLoungeAccess;
    private bool includesCatering;
    private string personalAssistant;
    private decimal serviceCharge;

    public string VipLoungeAccess
    {
        get { return vipLoungeAccess; }
        set { vipLoungeAccess = value; }
    }

    public bool IncludesCatering
    {
        get { return includesCatering; }
        set { includesCatering = value; }
    }

    public string PersonalAssistant
    {
        get { return personalAssistant; }
        set { personalAssistant = value; }
    }

    public decimal ServiceCharge
    {
        get { return serviceCharge; }
        set { serviceCharge = value; }
    }

    // Новые методы для VIP
    public void DisplayVIPBenefits()
    {
        Console.WriteLine($"VIP лаунж: {VipLoungeAccess}");
        Console.WriteLine($"Обслуживание: {(IncludesCatering ? "Включено" : "Не включено")}");
        Console.WriteLine($"Персональный ассистент: {PersonalAssistant}");
        Console.WriteLine($"Сервисный сбор: {ServiceCharge} руб.");
    }
    
    public override int CalculateLoyaltyPoints()
    {
        LoyaltyPoints = (int)((Price + ServiceCharge) * 0.2m); // Больше баллов за VIP
        return LoyaltyPoints;
    }
    
    public override void DisplayTicketInfo()
    {
        base.DisplayTicketInfo();
        DisplayVIPBenefits();
    }
    
    public VIPTicket(string playName, string director, DateTime eventDate,
                    string venue, decimal price, int actNumber)
        : base(playName, director, eventDate, venue, price, actNumber)
    {
        VipLoungeAccess = "Основной лаунж";
        IncludesCatering = true;
        PersonalAssistant = "Доступен";
        ServiceCharge = price * 0.15m;
    }
}

public class Customer : IValidatable
{
    private string name;
    private string email;
    private List<Ticket> purchasedTickets;

    // Новые атрибуты
    private string customerId;
    private string phoneNumber;
    private DateTime registrationDate;
    private decimal totalSpent;

    public string CustomerId
    {
        get { return customerId; }
        set { customerId = value; }
    }

    public string PhoneNumber
    {
        get { return phoneNumber; }
        set { phoneNumber = value; }
    }

    public DateTime RegistrationDate
    {
        get { return registrationDate; }
        set { registrationDate = value; }
    }

    public decimal TotalSpent
    {
        get { return totalSpent; }
        set { totalSpent = value; }
    }

    public string Name
    {
        get { return name; }
        set { name = value; }
    }

    public string Email
    {
        get { return email; }
        set { email = value; }
    }

    public List<Ticket> PurchasedTickets
    {
        get { return purchasedTickets; }
    }

    // Новые методы
    public bool Validate()
    {
        return !string.IsNullOrEmpty(Name) && !string.IsNullOrEmpty(Email) && Email.Contains("@");
    }
    
    public void UpdateSpending()
    {
        TotalSpent = 0;
        foreach (var ticket in purchasedTickets)
        {
            TotalSpent += ticket.Price;
        }
    }
    
    public int GetTotalLoyaltyPoints()
    {
        int totalPoints = 0;
        foreach (var ticket in purchasedTickets)
        {
            totalPoints += ticket.CalculateLoyaltyPoints();
        }
        return totalPoints;
    }
    
    public void DisplayCustomerStats()
    {
        Console.WriteLine($"\n=== СТАТИСТИКА КЛИЕНТА {Name} ===");
        Console.WriteLine($"ID: {CustomerId}");
        Console.WriteLine($"Зарегистрирован: {RegistrationDate:dd.MM.yyyy}");
        Console.WriteLine($"Всего потрачено: {TotalSpent} руб.");
        Console.WriteLine($"Накопленные баллы: {GetTotalLoyaltyPoints()}");
        Console.WriteLine($"Количество билетов: {purchasedTickets.Count}");
    }

    public Customer(string name, string email)
    {
        Name = name;
        Email = email;
        purchasedTickets = new List<Ticket>();
        CustomerId = Guid.NewGuid().ToString().Substring(0, 8);
        PhoneNumber = "";
        RegistrationDate = DateTime.Now;
        TotalSpent = 0;
    }

    public void PurchaseTicket(Ticket ticket)
    {
        purchasedTickets.Add(ticket);
        Console.WriteLine($"{Name} приобрел билет на {ticket.EventName}");
        UpdateSpending();
    }

    public void DisplayPurchasedTickets()
    {
        Console.WriteLine($"\n=== БИЛЕТЫ КЛИЕНТА {Name} ===");
        foreach (var ticket in purchasedTickets)
        {
            ticket.DisplayTicketInfo();
            Console.WriteLine("---");
        }
    }
}

public class TicketRepository : IValidatable
{
    private List<Ticket> tickets = new List<Ticket>();
    
    // Новые атрибуты
    private string repositoryName;
    private int maxCapacity;
    private DateTime lastBackup;
    
    public string RepositoryName
    {
        get { return repositoryName; }
        set { repositoryName = value; }
    }
    
    public int MaxCapacity
    {
        get { return maxCapacity; }
        set { maxCapacity = value; }
    }
    
    public DateTime LastBackup
    {
        get { return lastBackup; }
        set { lastBackup = value; }
    }
    
    // Новые методы
    public bool Validate()
    {
        return !string.IsNullOrEmpty(RepositoryName) && MaxCapacity > 0 && tickets.Count <= MaxCapacity;
    }
    
    public void BackupDatabase()
    {
        LastBackup = DateTime.Now;
        Console.WriteLine($"База данных {RepositoryName}备份完成 в {LastBackup:HH:mm:ss}");
    }
    
    public int GetTicketCountByType<T>() where T : Ticket
    {
        int count = 0;
        foreach (var ticket in tickets)
        {
            if (ticket is T)
                count++;
        }
        return count;
    }
    
    public void SaveToDatabase(Ticket ticket)
    {
        if (tickets.Count >= MaxCapacity && MaxCapacity > 0)
        {
            Console.WriteLine("Достигнута максимальная вместимость базы данных!");
            return;
        }
        
        tickets.Add(ticket);
        Console.WriteLine($"Билет '{ticket.EventName}' сохранен в базе данных {RepositoryName}");
    }
    
    public void DisplayAllTickets()
    {
        Console.WriteLine($"\n=== ВСЕ БИЛЕТЫ В БАЗЕ ДАННЫХ {RepositoryName} ===");
        foreach (var ticket in tickets)
        {
            ticket.DisplayTicketInfo();
            Console.WriteLine("---");
        }
    }

    public Ticket FindTicketByEventName(string eventName)
    {
        return tickets.Find(t => t.EventName.Equals(eventName));
    }
    
    public TicketRepository()
    {
        RepositoryName = "Основная база билетов";
        MaxCapacity = 1000;
        LastBackup = DateTime.Now;
    }
}

public class BookingSystem
{
    private TicketRepository repository;
    private List<Customer> customers;

    // Новые атрибуты
    private string systemVersion;
    private int totalBookings;
    private bool isSystemActive;
    
    public string SystemVersion
    {
        get { return systemVersion; }
        set { systemVersion = value; }
    }
    
    public int TotalBookings
    {
        get { return totalBookings; }
        set { totalBookings = value; }
    }
    
    public bool IsSystemActive
    {
        get { return isSystemActive; }
        set { isSystemActive = value; }
    }
    
    // Новые методы
    public void SystemDiagnostics()
    {
        Console.WriteLine($"\n=== ДИАГНОСТИКА СИСТЕМЫ ===");
        Console.WriteLine($"Версия: {SystemVersion}");
        Console.WriteLine($"Статус: {(IsSystemActive ? "Активна" : "Неактивна")}");
        Console.WriteLine($"Всего бронирований: {TotalBookings}");
        Console.WriteLine($"Количество клиентов: {customers.Count}");
        Console.WriteLine($"Заполненность базы: {repository.GetTicketCountByType<Ticket>()}/{(repository.MaxCapacity == 0 ? "∞" : repository.MaxCapacity.ToString())}");
    }
    
    public void GenerateReport()
    {
        Console.WriteLine($"\n=== ОТЧЕТ СИСТЕМЫ БРОНИРОВАНИЯ ===");
        Console.WriteLine($"Кино билеты: {repository.GetTicketCountByType<MovieTicket>()}");
        Console.WriteLine($"Концертные билеты: {repository.GetTicketCountByType<ConcertTicket>()}");
        Console.WriteLine($"Театральные билеты: {repository.GetTicketCountByType<TheaterTicket>()}");
        Console.WriteLine($"VIP билеты: {repository.GetTicketCountByType<VIPTicket>()}");
        
        decimal totalRevenue = 0;
        foreach (var customer in customers)
        {
            totalRevenue += customer.TotalSpent;
        }
        Console.WriteLine($"Общая выручка: {totalRevenue} руб.");
    }

    public BookingSystem()
    {
        repository = new TicketRepository();
        customers = new List<Customer>();
        SystemVersion = "2.0.0";
        TotalBookings = 0;
        IsSystemActive = true;
    }

    public void AddCustomer(Customer customer)
    {
        customers.Add(customer);
        Console.WriteLine($"Клиент {customer.Name} добавлен в систему");
    }

    public void ProcessBooking(Customer customer, Ticket ticket)
    {
        if (!IsSystemActive)
        {
            Console.WriteLine("Система бронирования временно недоступна!");
            return;
        }
        
        repository.SaveToDatabase(ticket);
        customer.PurchaseTicket(ticket);
        ticket.Reserve();
        TotalBookings++;
        
        // Верификация билета и расчет баллов
        ticket.VerifyTicket();
        int points = ticket.CalculateLoyaltyPoints();
        Console.WriteLine($"Начислено баллов лояльности: {points}");
    }

    public void DisplaySystemInfo()
    {
        Console.WriteLine($"\n=== ИНФОРМАЦИЯ СИСТЕМЫ БРОНИРОВАНИЯ ===");
        Console.WriteLine($"Количество клиентов: {customers.Count}");
        repository.DisplayAllTickets();
    }
}

class Program
{
    static void Main(string[] args)
    {
        // Создание системы бронирования
        BookingSystem bookingSystem = new BookingSystem();
        
        // Создание клиентов с ошибками в данных
        Customer customer1 = new Customer("", "invalid-email"); // Ошибка: пустое имя и неверный email
        Customer customer2 = new Customer(null, "maria@mail.com"); // Ошибка: null имя
        Customer customer3 = new Customer("Иван Петров", "ivan@mail.com"); // Корректный клиент
        
        bookingSystem.AddCustomer(customer1);
        bookingSystem.AddCustomer(customer2);
        bookingSystem.AddCustomer(customer3);
        
        // Создание билетов с ошибками в данных
        Ticket[] tickets = new Ticket[]
        {
            new MovieTicket("", new DateTime(2020, 12, 15, 19, 0, 0), // Ошибка: пустое название и прошедшая дата
                           "", -100, -1, ""), // Ошибка: отрицательная цена и номер зала
            new ConcertTicket("The Beatles", DateTime.Now.AddDays(10), "Стадион Лужники", 
                            2500, "Gold", true), // Корректный билет
            new TheaterTicket("Гамлет", "Сергей Женовач", DateTime.Now.AddDays(5), 
                            "МХТ им. Чехова", 1800, 3), // Корректный билет
            new VIPTicket("Щелкунчик", "Андрей Петров", DateTime.Now.AddDays(7), 
                        "Большой театр", 5000, 2) // VIP билет
        };
        
        // Настройка дополнительных атрибутов
        ((MovieTicket)tickets[0]).Genre = "Драма";
        ((MovieTicket)tickets[0]).DurationMinutes = 135;
        
        ((ConcertTicket)tickets[1]).MusicGenre = "Рок";
        ((ConcertTicket)tickets[1]).MeetAndGreet = true;
        
        ((TheaterTicket)tickets[2]).Playwright = "Уильям Шекспир";
        ((TheaterTicket)tickets[2]).HasLiveMusic = true;
        
        // Попытка обработки бронирований с ошибками
        try
        {
            bookingSystem.ProcessBooking(customer1, tickets[0]);
            bookingSystem.ProcessBooking(customer3, tickets[1]);
            bookingSystem.ProcessBooking(customer3, tickets[2]);
            bookingSystem.ProcessBooking(customer3, tickets[3]);
        }
        catch (Exception ex)
        {
            Console.WriteLine($"Ошибка: {ex.Message}");
        }
        
        // Демонстрация нового функционала
        Console.WriteLine("\n" + new string('=', 50));
        Console.WriteLine("ДЕМОНСТРАЦИЯ НОВОГО ФУНКЦИОНАЛА");
        Console.WriteLine(new string('=', 50));
        
        // Валидация
        Console.WriteLine($"Валидность клиента 1: {customer1.Validate()}");
        Console.WriteLine($"Валидность клиента 3: {customer3.Validate()}");
        Console.WriteLine($"Валидность билета 0: {tickets[0].Validate()}");
        Console.WriteLine($"Валидность билета 1: {tickets[1].Validate()}");
        
        // Экспорт данных
        Console.WriteLine($"\nЭкспорт JSON: {tickets[1].ExportToJson()}");
        Console.WriteLine($"Экспорт XML: {tickets[2].ExportToXml()}");
        
        // Статистика клиента
        customer3.DisplayCustomerStats();
        
        // Диагностика системы и отчет
        bookingSystem.SystemDiagnostics();
        bookingSystem.GenerateReport();
        
        // Резервное копирование
        bookingSystem.Repository.BackupDatabase();
        
        Console.ReadLine();
    }
    
    // Свойство для доступа к repository из BookingSystem
    private static TicketRepository Repository
    {
        get { return new TicketRepository(); }
    }
}
```
