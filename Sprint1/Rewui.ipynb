```polyglot-notebook
public class Product
{
    private string name;
    private decimal price;
    private int quantity;
    protected string category;
    internal string description;

    public Product(string name, decimal price, int quantity, string category, string description)
    {
        this.name = name;
        this.price = price;
        this.quantity = quantity;
        this.category = category;
        this.description = description;
    }

    // Геттеры и сеттеры
    public string Name
    {
        get { return name; }
        set { name = value; }
    }

    public decimal Price
    {
        get { return price; }
        set { price = value; }
    }

    public int Quantity
    {
        get { return quantity; }
        set { quantity = value; }
    }

    public string Category
    {
        get { return category; }
        set { category = value; }
    }

    public string Description
    {
        get { return description; }
        set { description = value; }
    }

    // Методы
    public decimal GetTotalPrice()
    {
        return price * quantity;
    }

    public void UpdateQuantity(int amount)
    {
        quantity += amount;
        if (quantity < 0) quantity = 0;
    }

    public string GetProductInfo()
    {
        return $"Товар: {name}, Цена: {price:C}, Количество: {quantity}, Категория: {category}, Описание: {description}";
    }
}

// Класс Продавец (Seller)
public class Seller
{
    private string name;
    private string employeeId;
    protected decimal salary;
    internal string contactInfo;
    private List<Product> products;

    public Seller(string name, string employeeId, decimal salary, string contactInfo)
    {
        this.name = name;
        this.employeeId = employeeId;
        this.salary = salary;
        this.contactInfo = contactInfo;
        this.products = new List<Product>();
    }

    // Геттеры и сеттеры
    public string Name
    {
        get { return name; }
        set { name = value; }
    }

    public string EmployeeId
    {
        get { return employeeId; }
        set { employeeId = value; }
    }

    public decimal Salary
    {
        get { return salary; }
        set { salary = value; }
    }

    public string ContactInfo
    {
        get { return contactInfo; }
        set { contactInfo = value; }
    }

    // Методы
    public void AddProduct(Product product)
    {
        products.Add(product);
        Console.WriteLine($"Товар '{product.Name}' добавлен продавцу {name}");
    }

    public void SellProduct(Product product, int quantity)
    {
        if (product.Quantity >= quantity)
        {
            product.UpdateQuantity(-quantity);
            Console.WriteLine($"Продано {quantity} единиц товара '{product.Name}' продавцом {name}");
        }
        else
        {
            Console.WriteLine($"Недостаточно товара '{product.Name}' для продажи. Доступно: {product.Quantity}");
        }
    }

    public void ListProducts()
    {
        Console.WriteLine($"Товары продавца {name}:");
        foreach (var product in products)
        {
            Console.WriteLine($"- {product.Name} (Количество: {product.Quantity})");
        }
    }
}

// Класс Магазин (Store)
public class Store
{
    private string storeName;
    private string location;
    public string storeHours;
    private List<Seller> sellers;
    private List<Product> products;

    public Store(string storeName, string location, string storeHours)
    {
        this.storeName = storeName;
        this.location = location;
        this.storeHours = storeHours;
        this.sellers = new List<Seller>();
        this.products = new List<Product>();
    }

    // Геттеры и сеттеры
    public string StoreName
    {
        get { return storeName; }
        set { storeName = value; }
    }

    public string Location
    {
        get { return location; }
        set { location = value; }
    }

    public string StoreHours
    {
        get { return storeHours; }
        set { storeHours = value; }
    }

    // Методы
    public void AddSeller(Seller seller)
    {
        sellers.Add(seller);
        Console.WriteLine($"Продавец {seller.Name} добавлен в магазин {storeName}");
    }

    public void AddProduct(Product product)
    {
        products.Add(product);
        Console.WriteLine($"Товар '{product.Name}' добавлен в магазин {storeName}");
    }

    public void ListProducts()
    {
        Console.WriteLine($"Товары в магазине {storeName}:");
        foreach (var product in products)
        {
            Console.WriteLine($"- {product.GetProductInfo()}");
        }
    }

    public string GetStoreInfo()
    {
        return $"Магазин: {storeName}, Адрес: {location}, Часы работы: {storeHours}, Продавцов: {sellers.Count}, Товаров: {products.Count}";
    }

    public void ListSellers()
    {
        Console.WriteLine($"Продавцы магазина {storeName}:");
        foreach (var seller in sellers)
        {
            Console.WriteLine($"- {seller.Name} (ID: {seller.EmployeeId})");
        }
    }
}
Console.WriteLine("=== Демонстрация работы с магазином ===\n");

        // Создание товаров
        Product laptop = new Product("Ноутбук", 50000m, 10, "Электроника", "Игровой ноутбук");
        Product phone = new Product("Смартфон", 25000m, 20, "Электроника", "Флагманский смартфон");
        Product book = new Product("Книга", 500m, 50, "Книги", "Художественная литература");

        // Создание продавцов
        Seller seller1 = new Seller("Иван Иванов", "EMP001", 40000m, "ivan@store.com");
        Seller seller2 = new Seller("Мария Петрова", "EMP002", 45000m, "maria@store.com");

        // Создание магазина
        Store myStore = new Store("ТехноМир", "ул. Центральная, 123", "9:00-21:00");

        // Добавление товаров в магазин
        myStore.AddProduct(laptop);
        myStore.AddProduct(phone);
        myStore.AddProduct(book);

        // Добавление продавцов в магазин
        myStore.AddSeller(seller1);
        myStore.AddSeller(seller2);

        // Добавление товаров продавцам
        seller1.AddProduct(laptop);
        seller1.AddProduct(phone);
        seller2.AddProduct(book);

        Console.WriteLine("\n" + myStore.GetStoreInfo());

        Console.WriteLine("\n=== Информация о товарах ===");
        myStore.ListProducts();

        Console.WriteLine("\n=== Продавцы ===");
        myStore.ListSellers();

        Console.WriteLine("\n=== Товары продавцов ===");
        seller1.ListProducts();
        seller2.ListProducts();

        Console.WriteLine("\n=== Демонстрация продаж ===");
        seller1.SellProduct(laptop, 2);
        seller2.SellProduct(book, 5);
        seller1.SellProduct(phone, 25); // Попытка продать больше чем есть

        Console.WriteLine("\n=== Обновленная информация о товарах ===");
        myStore.ListProducts();

        Console.WriteLine("\n=== Общая стоимость товаров ===");
        Console.WriteLine($"Ноутбуки: {laptop.GetTotalPrice():C}");
        Console.WriteLine($"Смартфоны: {phone.GetTotalPrice():C}");
        Console.WriteLine($"Книги: {book.GetTotalPrice():C}");
```
